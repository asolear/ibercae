```python
# ///////////////////////////////////////////////
#  definir en las tablas las variables
from numpy import *
import pandas as pd
import datetime

pd.options.mode.chained_assignment = None  # default='warn'
import matplotlib.pyplot as plt
import base64
from io import BytesIO
```

```python
# 
df=pd.read_csv("datos/reactiva.csv", header=[0])
aa=df.describe()
# df['tercio1']=df.Kvar1/df.kWh1*100
# df['tercio2']=df.Kvar2/df.kWh2*100

df['exceso1']=where(df.Kvar1-df.kWh1/3 < 0,0,df.Kvar1-df.kWh1/3 )
df['exceso2']=where(df.Kvar2-df.kWh2/3 < 0,0,df.Kvar2-df.kWh2/3 )
# df
df['cfi1']=cos(arctan(df.Kvar1/df.kWh1))
df['cfi2']=cos(arctan(df.Kvar2/df.kWh2))
# df
df['rec1']= where(df.cfi1 >= 0.95,0,where(df.cfi1 <= 0.80,0.062332 ,0.041554 ))
df['rec2']= where(df.cfi2 >= 0.95,0,where(df.cfi2 <= 0.80,0.062332 ,0.041554 ))

df['rect1']=df.exceso1*df.rec1
df['rect2']=df.exceso2*df.rec2
recargo_total=df.sum().rect1+df.sum().rect2
recargo_total


# df
```
