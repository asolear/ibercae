```python
import requests as requests
import json
# https://ec.europa.eu/jrc/en/PVGIS/docs/noninteractive
# entradas proporcionadas
ciudad = "Badajoz"
lat=38.376
lon=-6.967
# salidas de simulacion

r = requests.get("https://re.jrc.ec.europa.eu/api/DRcalc?lat="+str(lat)+"&lon="+str(lon)+"&month=3&global=1&outputformat=json")
data = r.json()
# data #para ver los datos a parsear
j=data["outputs"]["daily_profile"]
kf= json.dumps(j, indent=2)
df=pd.read_json(kf, orient='records') 
df.plot(kind='bar',x='time',y=['G(i)']);
```
