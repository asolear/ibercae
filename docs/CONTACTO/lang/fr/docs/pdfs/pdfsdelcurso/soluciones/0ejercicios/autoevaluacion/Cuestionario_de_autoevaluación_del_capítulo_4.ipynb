Seleccione una:

b. a las 11:30

```python
index=["11:00","11:30"]

df=pd.DataFrame(index=index)
dd=pd.DataFrame()
dd['ud']=['$m^3$h','C']

# autoevaluacion
df['caudal']=[1.2,2]
df['Timpulsion']=[7.1,7.5]
df['Tretorno']=[11.8,12]
df['electrica']=[2.2,3.1]
# ejemplo p-184
# df
# dd
# potencia térmica que pasa por la tubería.

# Pot (kW) = caudal (litros/minuto) x 60 min/hora x DT (oC) / 860 kcalh/kW)
df['termica']=df.caudal*60*(df.Tretorno-df.Timpulsion)/860

df['EER']=df.termica/df.electrica

df
```
