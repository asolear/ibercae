El primer  paso para poder analizar los costes eléctricos es saber que componentes  tienes en una factura eléctrica y tendremos que dividir el coste de las  facturas del edificio auditado en todos estos conceptos. 

- Término de potencia (€/kW) es el abono de los derechos de uso de la infraestructura necesaria para tener la potencia contratada. El 100% son costes regulados. Se paga por kW contratado, lo utilices o no.
- Término de energía (€/kWh) es el precio que el consumidor paga por la energía que consume. Una parte son peajes de acceso y son regulados.
- Recargo por exceso de potencia: si sobrepasamos por encima del 105% de la potencia contratada tendremos un recargo.
- Recargo por energía reactiva:si la energía reactiva sobrepasa el 33% de la energía activa tendremos un recargo.
- Impuesto eléctrico:impuesto de 5,1127% sobre el total del coste de la factura.
- Alquiler de equipos de medida:coste de alquilar el equipo de medida y así evitar la inversión de comprarlo.

## Optimizar potencia contratada.
Para el cálculo del término de potencia en los contratos 3.X, se mira la potencia máxima que se ha registrado para cada uno de los tres periodos del mes de facturación. Una vez tenemos  la potencia máxima registrada se aplican los siguiente supuestos para  cada una:

$$
  P_{medida} <  0.85\ P_{contratada} \Rightarrow P_{facturada}=P_{contratada}
$$

$$
0.85\ P_{contratada} \leq P_{medida} \leq 1.05\ P_{contratada}\Rightarrow P_{facturada}=P_{medida}
$$

$$
1.05\ P_{contratada} < P_{medida} \Rightarrow P_{facturada}=P_{medida} + 2(P_{medida}-1.05\ P_{contratada})
$$
